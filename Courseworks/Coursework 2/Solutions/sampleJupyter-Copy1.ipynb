We create a brand new txt file from the Jupyter notebook itself.

```python
from random import *
```

```python
file=open("table1.txt","w")
for x in range(5):
    file.write("{:.20f} \t {:.20f} \n".format(random(),random()))

file.close()
```

```python
run "refdefpython.py" table1.txt 100000
```

```python
file=open("table2.txt","w")
for x in range(6):
    file.write("{:.20f} \t {:.20f} \n".format(random(),random()))

file.close()
```

```python
run "refdefpython.py" table2.txt 100000
```

```python
file=open("table3.txt","w")
for x in range(3):
    file.write("{:.20f} \t {:.20f} \n".format(random(),random()))
    file.write("{:.20f} \t {:.20f} \n".format(-random(),random()))
file.close()
```

```python
run "refdefpython.py" table3.txt 100000
```
