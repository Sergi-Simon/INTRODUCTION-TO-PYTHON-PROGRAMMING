If we increase the number of points the difference between the recursive and non-recursive functions becomes evident. If you do this in your Jupyter notebook (which is entirely optional), you don't need to copy your new data files table0.txt, etc into the zip file because the Jupyter notebook should be able to do that on its own.

```python
with open("table.txt") as file:
    with open("table0.txt", "w") as file0:
        for line in file:
            file0.write(line)
file0.close()
file.close()
```

```python
file=open("table0.txt","a")
file.write("{:.20f} \t {:.20f} \n".format(8.1,1))
file.write("{:.20f} \t {:.20f} \n".format(-2,-1))
file.write("{:.20f} \t {:.20f} \n".format(-3, 0))
file.close()
```

```python
run "interpolation.py" table0.txt 100000
```

We keep adding points:

```python
file=open("table0.txt","a")
file.write("{:.20f} \t {:.20f} \n".format(-9,1))
file.write("{:.20f} \t {:.20f} \n".format(-10,2))
file.write("{:.20f} \t {:.20f} \n".format(10, 1))
file.close()
```

```python
run "interpolation.py" table0.txt 100000
```

We create a brand new txt file from the Jupyter notebook itself.

```python
from random import *
```

```python
file=open("table1.txt","w")
for x in range(5):
    file.write("{:.20f} \t {:.20f} \n".format(random(),random()))

file.close()
```

```python
run "interpolation.py" table1.txt 100000
```

```python
file=open("table2.txt","w")
for x in range(10):
    file.write("{:.20f} \t {:.20f} \n".format(random(),random()))

file.close()
```

```python
run "interpolation.py" table2.txt 100000
```

```python
file=open("table3.txt","w")
for x in range(10):
    file.write("{:.20f} \t {:.20f} \n".format(random(),random()))
    file.write("{:.20f} \t {:.20f} \n".format(-random(),random()))
file.close()
```

```python
run "interpolation.py" table3.txt 100000
```
